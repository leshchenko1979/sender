```python
lines = msg.text.split("\n")

# –ù–∞–π—Ç–∏ –∏–Ω–¥–µ–∫—Å—ã —Å—Ç—Ä–æ–∫ —Å–æ —Å–ª–æ–≤–∞–º–∏ "–ü–æ–¥–ø–∏—Å–∞–ª–∏—Å—å" –∏ "–û—Ç–ø–∏—Å–∞–ª–∏—Å—å"
for i, s in enumerate(lines):
    if "–ü–æ–¥–ø–∏—Å–∞–ª–∏—Å—å:" in s:
        sub_index = i
    if "–û—Ç–ø–∏—Å–∞–ª–∏—Å—å:" in s:
        unsub_index = i
        break


def scrape_usernames(lines: list[str]) -> set[str]:
    usernames = set()
    for s in lines:
        try:
            found = next(re.finditer(r"@\S+", s))
            usernames.add(found.group(0))
        except StopIteration:
            continue

    return usernames


subscribers = scrape_usernames(lines[sub_index:unsub_index])
unsubscribers = scrape_usernames(lines[unsub_index:])

subscribers, unsubscribers
```

```python
async with acc.session(revalidate=True):
    app: pyrogram.Client = acc.app
    for u in subscribers:
        await app.send_message(
            u,
            "–ü—Ä–∏–≤–µ—Ç! –≠—Ç–æ –ê–ª–µ–∫—Å–µ–π –õ–µ—â–µ–Ω–∫–æ.\n\n"
            '–í—ã –Ω–µ–¥–∞–≤–Ω–æ –ø–æ–¥–ø–∏—Å–∞–ª–∏—Å—å –Ω–∞ –º–æ–π –∫–∞–Ω–∞–ª "–ò–Ω–≤–µ—Å—Ç–∏—Ü–∏–∏ –≤ —Ä–µ–¥–µ–≤–µ–ª–æ–ø–º–µ–Ω—Ç".\n\n'
            "–°–ø–∞—Å–∏–±–æ –∏ –¥–æ–±—Ä–æ –ø–æ–∂–∞–ª–æ–≤–∞—Ç—å! :)\n\n"
            "–°–∫–∞–∂–∏—Ç–µ, —á—Ç–æ –≤–∞—Å –ø—Ä–∏–≤–ª–µ–∫–ª–æ –∏ —á—Ç–æ –±—ã –≤—ã —Ö–æ—Ç–µ–ª–∏ –≤–∏–¥–µ—Ç—å –Ω–∞ –∫–∞–Ω–∞–ª–µ?",
        )
    for u in unsubscribers:
        await app.send_message(
            u,
            "–ü—Ä–∏–≤–µ—Ç! –≠—Ç–æ –ê–ª–µ–∫—Å–µ–π –õ–µ—â–µ–Ω–∫–æ.\n\n"
            '–í—ã –Ω–µ–¥–∞–≤–Ω–æ –æ—Ç–ø–∏—Å–∞–ª–∏—Å—å –æ—Ç –º–æ–µ–≥–æ –∫–∞–Ω–∞–ª–∞ "–ò–Ω–≤–µ—Å—Ç–∏—Ü–∏–∏ –≤ —Ä–µ–¥–µ–≤–µ–ª–æ–ø–º–µ–Ω—Ç".\n\n'
            "–Ø –≥—Ä—É—â—É. :(\n\n"
            "–ü–æ–¥—Å–∫–∞–∂–∏—Ç–µ, –∞ –ø–æ—á–µ–º—É –∫–∞–Ω–∞–ª –æ–∫–∞–∑–∞–ª—Å—è –±–µ—Å–ø–æ–ª–µ–∑–Ω—ã–º?\n\n"
            "–ß–µ–≥–æ –±—ã –≤—ã —Ö–æ—Ç–µ–ª–∏ –≤–∏–¥–µ—Ç—å –±–æ–ª—å—à–µ, –∞ —á–µ–≥–æ - –º–µ–Ω—å—à–µ?",
        )
```
